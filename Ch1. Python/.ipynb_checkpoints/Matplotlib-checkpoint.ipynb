```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1) #以0.1为单位，生成0-6的数
print(x)
y = np.sin(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle = "--",label = "cos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
